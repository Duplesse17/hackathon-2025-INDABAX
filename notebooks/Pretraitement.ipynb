```python
#determinon les cordonner des quartier 
if 'Latitude' not in df.columns or 'Longitude' not in df.columns:
    print("=> Géocodage en cours...")

    geolocator = Nominatim(user_agent="dash_map_app")

    quartiers_uniques = df[['Quartier de Résidence']].drop_duplicates()

    latitudes = []
    longitudes = []

    for quartier in quartiers_uniques['Quartier de Résidence']:
        try:
            location = geolocator.geocode(f"{quartier}, Cameroun")
            if location:
                latitudes.append(location.latitude)
                longitudes.append(location.longitude)
            else:
                latitudes.append(None)
                longitudes.append(None)
            time.sleep(1.2)  # Pause de 1.2 seconde pour éviter de surcharger le serveur
        except Exception as e:
            latitudes.append(None)
            longitudes.append(None)
            print(f"Erreur sur {quartier} : {e}")

    quartiers_uniques['Latitude'] = latitudes
    quartiers_uniques['Longitude'] = longitudes

    # Fusion avec les données d'origine
    df = df.merge(quartiers_uniques, on='Quartier de Résidence', how='left')

    # Sauvegarde
    # df.to_excel('donneurs_geocode.xlsx', index=False)
else:
    #df = pd.read_excel('donneurs_geocode.xlsx')
    print('colonne lattitude et longitude existe dejas')

print("✅ Géocodage terminé !")
```

```python
# Vérifier si les colonnes existent, sinon les créer
if 'Latitude' not in df.columns or 'Longitude' not in df.columns:
    df['Latitude'] = None
    df['Longitude'] = None

# Filtrer uniquement les quartiers sans coordonnées
quartiers_sans_coord = df[df['Latitude'].isna() | df['Longitude'].isna()][['Quartier de Résidence']].drop_duplicates().reset_index()

if not quartiers_sans_coord.empty:
    print("=> Géocodage en cours...")

    geolocator = Nominatim(user_agent="dash_map_app")
    
    for idx in quartiers_sans_coord.index:
        quartier = quartiers_sans_coord.loc[idx, 'Quartier de Résidence']
        try:
            location = geolocator.geocode(f"{quartier}, Cameroun")
            if location:
                quartiers_sans_coord.loc[idx, 'Latitude'] = location.latitude
                quartiers_sans_coord.loc[idx, 'Longitude'] = location.longitude
                print(f"{quartier} : {location.latitude}, {location.longitude}")
            else:
                quartiers_sans_coord.loc[idx, 'Latitude'] = None
                quartiers_sans_coord.loc[idx, 'Longitude'] = None
                print(f"{quartier} : {location.latitude}, {location.longitude}")
            time.sleep(1.2)  # Pause pour éviter de surcharger l'API
        except Exception as e:
            quartiers_sans_coord.loc[idx, 'Latitude'] = None
            quartiers_sans_coord.loc[idx, 'Longitude'] = None
            print(f"Erreur sur {quartier} : {e}")

    # Fusionner uniquement les nouvelles coordonnées sans écraser les anciennes
    df.update(quartiers_sans_coord.set_index('index'))

    print("✅ Géocodage terminé et mis à jour !")
else:
    print("✅ Tous les quartiers ont déjà des coordonnées.")
```

```python
# -------------------------------------
# ETAPE 2 : Correction des Arrondissements Non Précisés
# -------------------------------------

print("🔎 Correction des Arrondissements...")

geolocator = Nominatim(user_agent="dash_map_app_reverse")

# On filtre uniquement les lignes avec "Douala (non précisé)"
mask_non_precise = df['Arrondissement de résidence'].str.strip().str.capitalize() == 'Douala (non précisé )'

# Pour éviter des erreurs si pas de lignes concernées
if mask_non_precise.sum() > 0:
    for idx, row in df[mask_non_precise].iterrows():
        lat = row['Latitude']
        lon = row['Longitude']
        
        if pd.notnull(lat) and pd.notnull(lon):
            try:
                # Reverse geocoding
                location = geolocator.reverse((lat, lon), exactly_one=True, language='fr')
                time.sleep(1)

                if location and 'address' in location.raw:
                    address = location.raw['address']

                    # Recherche d'un niveau d'arrondissement ou de quartier administratif
                    arrondissement = address.get('suburb') or address.get('city_district') or address.get('county')

                    if arrondissement:
                        df.at[idx, 'Arrondissement de résidence'] = arrondissement
                        print(f"✅ Arrondissement mis à jour : {arrondissement}")
                    else:
                        print(f"❗️ Arrondissement non trouvé pour les coordonnées : {lat}, {lon}")
                else:
                    print(f"❗️ Pas de réponse valide pour les coordonnées : {lat}, {lon}")

            except Exception as e:
                print(f"Erreur reverse geocoding à l'index {idx} : {e}")

else:
    print("✅ Aucun enregistrement 'Douala (non précisé)' trouvé.")
```

### Travail sur le dataset des donneurs

```python
# Lire la feuille 'Donneurs 2019' du fichier source
df_donneurs_2019 = pd.read_excel('../data/Updated_Challenge_dataset.xlsx', sheet_name='Donneurs 2019')

# Supprimer la colonne 'Horodateur'
df_donneurs_2019 = df_donneurs_2019.drop(columns=['Horodateur'])

# -------------------------------
# Nettoyage de la colonne Age
# -------------------------------

# Renommer les colonnes pour enlever les espaces superflus
df_donneurs_2019.rename(columns=lambda x: x.strip(), inplace=True)

# Convertir 'Age' en float (forcer les erreurs à NaN)
df_donneurs_2019['Age'] = pd.to_numeric(df_donneurs_2019['Age'], errors='coerce')

# Afficher les types après conversion
print("\n✅ Types des colonnes après conversion :")
print(df_donneurs_2019.dtypes)

# Remplacer les 0 par NaN (on considère qu'un âge à 0 est invalide)
df_donneurs_2019['Age'].replace(0, np.nan)

# Calculer la moyenne en ignorant les NaN
age_moyen = df_donneurs_2019['Age'].mean()

# Remplacer les NaN par la moyenne
df_donneurs_2019['Age'].fillna(age_moyen)

print(f"\n✅ Valeurs manquantes ou nulles dans 'Age' remplacées par la moyenne : {round(age_moyen, 2)}")

# -------------------------------
# Supprimer les lignes où 'Type de donation' est vide
# -------------------------------

# Supprimer les lignes avec NaN dans 'Type de donation'
df_donneurs_2019 = df_donneurs_2019.dropna(subset=['Type de donation'])

print(f"\n✅ Nombre total de lignes après suppression des 'Type de donation' vides : {df_donneurs_2019.shape[0]}")

# -------------------------------
# Vérifications finales
# -------------------------------

# Compter les valeurs nulles par colonne (après nettoyage)
print("\n✅ Nombre de valeurs nulles par colonne après nettoyage :")
print(df_donneurs_2019.isnull().sum())

# Voir un aperçu des données nettoyées
print("\n✅ Aperçu des premières lignes :")
print(df_donneurs_2019.head())

# Nombre total de donneurs non éligibles
total_non_eligibles = len(df_donneurs_2019)

# Répartition par sexe
repartition_sexe = df_donneurs_2019['Sexe'].value_counts()

# Âge moyen
age_moyen = df_donneurs_2019['Age'].mean()

# Type de donation dominant
donation_dominante = df_donneurs_2019['Type de donation'].mode()[0]

# Groupe sanguin dominant
groupe_sanguin_dominant = df_donneurs_2019['Groupe Sanguin ABO / Rhesus'].mode()[0]

# Phénotype dominant
phenotype_dominant = df_donneurs_2019['Phenotype'].mode()[0]

print(f"Total non éligibles : {total_non_eligibles}")
print(f"Répartition sexe : \n{repartition_sexe}")
print(f"Âge moyen : {age_moyen:.1f} ans")
print(f"Type de donation dominant : {donation_dominante}")
print(f"Groupe sanguin dominant : {groupe_sanguin_dominant}")
print(f"Phénotype dominant : {phenotype_dominant}")


# Enregistrer le fichier nettoyé si besoin
df_donneurs_2019.to_excel('../data/donneurs_2019_nettoye.xlsx', index=False)
print("\n✅ Fichier nettoyé sauvegardé sous '../data/donneurs_2019_nettoye.xlsx' !")
```
