```python
#determinon les cordonner des quartier 
if 'Latitude' not in df.columns or 'Longitude' not in df.columns:
    print("=> GÃ©ocodage en cours...")

    geolocator = Nominatim(user_agent="dash_map_app")

    quartiers_uniques = df[['Quartier de RÃ©sidence']].drop_duplicates()

    latitudes = []
    longitudes = []

    for quartier in quartiers_uniques['Quartier de RÃ©sidence']:
        try:
            location = geolocator.geocode(f"{quartier}, Cameroun")
            if location:
                latitudes.append(location.latitude)
                longitudes.append(location.longitude)
            else:
                latitudes.append(None)
                longitudes.append(None)
            time.sleep(1.2)  # Pause de 1.2 seconde pour Ã©viter de surcharger le serveur
        except Exception as e:
            latitudes.append(None)
            longitudes.append(None)
            print(f"Erreur sur {quartier} : {e}")

    quartiers_uniques['Latitude'] = latitudes
    quartiers_uniques['Longitude'] = longitudes

    # Fusion avec les donnÃ©es d'origine
    df = df.merge(quartiers_uniques, on='Quartier de RÃ©sidence', how='left')

    # Sauvegarde
    # df.to_excel('donneurs_geocode.xlsx', index=False)
else:
    #df = pd.read_excel('donneurs_geocode.xlsx')
    print('colonne lattitude et longitude existe dejas')

print("âœ… GÃ©ocodage terminÃ© !")
```

```python
# VÃ©rifier si les colonnes existent, sinon les crÃ©er
if 'Latitude' not in df.columns or 'Longitude' not in df.columns:
    df['Latitude'] = None
    df['Longitude'] = None

# Filtrer uniquement les quartiers sans coordonnÃ©es
quartiers_sans_coord = df[df['Latitude'].isna() | df['Longitude'].isna()][['Quartier de RÃ©sidence']].drop_duplicates().reset_index()

if not quartiers_sans_coord.empty:
    print("=> GÃ©ocodage en cours...")

    geolocator = Nominatim(user_agent="dash_map_app")
    
    for idx in quartiers_sans_coord.index:
        quartier = quartiers_sans_coord.loc[idx, 'Quartier de RÃ©sidence']
        try:
            location = geolocator.geocode(f"{quartier}, Cameroun")
            if location:
                quartiers_sans_coord.loc[idx, 'Latitude'] = location.latitude
                quartiers_sans_coord.loc[idx, 'Longitude'] = location.longitude
                print(f"{quartier} : {location.latitude}, {location.longitude}")
            else:
                quartiers_sans_coord.loc[idx, 'Latitude'] = None
                quartiers_sans_coord.loc[idx, 'Longitude'] = None
                print(f"{quartier} : {location.latitude}, {location.longitude}")
            time.sleep(1.2)  # Pause pour Ã©viter de surcharger l'API
        except Exception as e:
            quartiers_sans_coord.loc[idx, 'Latitude'] = None
            quartiers_sans_coord.loc[idx, 'Longitude'] = None
            print(f"Erreur sur {quartier} : {e}")

    # Fusionner uniquement les nouvelles coordonnÃ©es sans Ã©craser les anciennes
    df.update(quartiers_sans_coord.set_index('index'))

    print("âœ… GÃ©ocodage terminÃ© et mis Ã  jour !")
else:
    print("âœ… Tous les quartiers ont dÃ©jÃ  des coordonnÃ©es.")
```

```python
# -------------------------------------
# ETAPE 2 : Correction des Arrondissements Non PrÃ©cisÃ©s
# -------------------------------------

print("ğŸ” Correction des Arrondissements...")

geolocator = Nominatim(user_agent="dash_map_app_reverse")

# On filtre uniquement les lignes avec "Douala (non prÃ©cisÃ©)"
mask_non_precise = df['Arrondissement de rÃ©sidence'].str.strip().str.capitalize() == 'Douala (non prÃ©cisÃ© )'

# Pour Ã©viter des erreurs si pas de lignes concernÃ©es
if mask_non_precise.sum() > 0:
    for idx, row in df[mask_non_precise].iterrows():
        lat = row['Latitude']
        lon = row['Longitude']
        
        if pd.notnull(lat) and pd.notnull(lon):
            try:
                # Reverse geocoding
                location = geolocator.reverse((lat, lon), exactly_one=True, language='fr')
                time.sleep(1)

                if location and 'address' in location.raw:
                    address = location.raw['address']

                    # Recherche d'un niveau d'arrondissement ou de quartier administratif
                    arrondissement = address.get('suburb') or address.get('city_district') or address.get('county')

                    if arrondissement:
                        df.at[idx, 'Arrondissement de rÃ©sidence'] = arrondissement
                        print(f"âœ… Arrondissement mis Ã  jour : {arrondissement}")
                    else:
                        print(f"â—ï¸ Arrondissement non trouvÃ© pour les coordonnÃ©es : {lat}, {lon}")
                else:
                    print(f"â—ï¸ Pas de rÃ©ponse valide pour les coordonnÃ©es : {lat}, {lon}")

            except Exception as e:
                print(f"Erreur reverse geocoding Ã  l'index {idx} : {e}")

else:
    print("âœ… Aucun enregistrement 'Douala (non prÃ©cisÃ©)' trouvÃ©.")
```

```python
# Lire la feuille 'Donneurs 2019' du fichier source
df_donneurs_2019 = pd.read_excel('../data/Updated_Challenge_dataset.xlsx', sheet_name='Donneurs 2019')

# Supprimer la colonne 'Horodateur'
df_donneurs_2019 = df_donneurs_2019.drop(columns=['Horodateur'])

# -------------------------------
# Nettoyage de la colonne Age
# -------------------------------

# Renommer les colonnes pour enlever les espaces superflus
df_donneurs_2019.rename(columns=lambda x: x.strip(), inplace=True)

# Convertir 'Age' en float (forcer les erreurs Ã  NaN)
df_donneurs_2019['Age'] = pd.to_numeric(df_donneurs_2019['Age'], errors='coerce')

# Afficher les types aprÃ¨s conversion
print("\nâœ… Types des colonnes aprÃ¨s conversion :")
print(df_donneurs_2019.dtypes)

# Remplacer les 0 par NaN (on considÃ¨re qu'un Ã¢ge Ã  0 est invalide)
df_donneurs_2019['Age'].replace(0, np.nan)

# Calculer la moyenne en ignorant les NaN
age_moyen = df_donneurs_2019['Age'].mean()

# Remplacer les NaN par la moyenne
df_donneurs_2019['Age'].fillna(age_moyen)

print(f"\nâœ… Valeurs manquantes ou nulles dans 'Age' remplacÃ©es par la moyenne : {round(age_moyen, 2)}")

# -------------------------------
# Supprimer les lignes oÃ¹ 'Type de donation' est vide
# -------------------------------

# Supprimer les lignes avec NaN dans 'Type de donation'
df_donneurs_2019 = df_donneurs_2019.dropna(subset=['Type de donation'])

print(f"\nâœ… Nombre total de lignes aprÃ¨s suppression des 'Type de donation' vides : {df_donneurs_2019.shape[0]}")

# -------------------------------
# VÃ©rifications finales
# -------------------------------

# Compter les valeurs nulles par colonne (aprÃ¨s nettoyage)
print("\nâœ… Nombre de valeurs nulles par colonne aprÃ¨s nettoyage :")
print(df_donneurs_2019.isnull().sum())

# Voir un aperÃ§u des donnÃ©es nettoyÃ©es
print("\nâœ… AperÃ§u des premiÃ¨res lignes :")
print(df_donneurs_2019.head())

# Nombre total de donneurs non Ã©ligibles
total_non_eligibles = len(df_donneurs_2019)

# RÃ©partition par sexe
repartition_sexe = df_donneurs_2019['Sexe'].value_counts()

# Ã‚ge moyen
age_moyen = df_donneurs_2019['Age'].mean()

# Type de donation dominant
donation_dominante = df_donneurs_2019['Type de donation'].mode()[0]

# Groupe sanguin dominant
groupe_sanguin_dominant = df_donneurs_2019['Groupe Sanguin ABO / Rhesus'].mode()[0]

# PhÃ©notype dominant
phenotype_dominant = df_donneurs_2019['Phenotype'].mode()[0]

print(f"Total non Ã©ligibles : {total_non_eligibles}")
print(f"RÃ©partition sexe : \n{repartition_sexe}")
print(f"Ã‚ge moyen : {age_moyen:.1f} ans")
print(f"Type de donation dominant : {donation_dominante}")
print(f"Groupe sanguin dominant : {groupe_sanguin_dominant}")
print(f"PhÃ©notype dominant : {phenotype_dominant}")


# Enregistrer le fichier nettoyÃ© si besoin
df_donneurs_2019.to_excel('../data/donneurs_2019_nettoye.xlsx', index=False)
print("\nâœ… Fichier nettoyÃ© sauvegardÃ© sous '../data/donneurs_2019_nettoye.xlsx' !")
```
